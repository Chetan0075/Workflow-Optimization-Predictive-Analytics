```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

delays = pd.read_csv('D:/Mega Project Data Analysis/sql/exported_data/delays.csv')
alloc  = pd.read_csv('D:/Mega Project Data Analysis/sql/exported_data/resourse_allocation.csv')
tasks  = pd.read_csv('D:/Mega Project Data Analysis/sql/exported_data/tasks.csv')

df = tasks.merge(alloc, on='task_id', how='left')
df['delayed'] = df['task_id'].isin(delays['task_id']).astype(int)

features = ['estimated_hours','allocated_hours','utilization_pct']
X = df[features]
y = df['delayed']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

model = RandomForestClassifier()
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
from pathlib import Path
import joblib

model_dir = Path('models')
model_dir.mkdir(parents=True, exist_ok=True)

joblib.dump(model, model_dir / 'delay_predictor.pkl')
```
